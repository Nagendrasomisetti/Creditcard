```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside the notebook
%matplotlib inline

# Load data
df = pd.read_csv("../data/creditcard.csv")  # Adjust if your path is different

# Show basic info
df.info()
df.head()
```

```python
# Check class distribution
class_counts = df['Class'].value_counts()
print(class_counts)

# Plot class distribution
plt.figure(figsize=(6,4))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title("Class Distribution (0: Normal, 1: Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# Show imbalance in %
fraud_pct = round(class_counts[1] / class_counts.sum() * 100, 4)
print(f"Fraudulent Transactions: {fraud_pct}%")
```

```python
# Compute correlation
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(corr, cmap='coolwarm_r', annot=False)
plt.title("Correlation Heatmap")
plt.show()

# Check correlation with 'Class'
corr_target = corr['Class'].sort_values(ascending=False)
print("Top correlated features with Class:")
print(corr_target.head(10))
```

```python
top_features = corr_target[1:4].index.tolist()  # Skip 'Class' itself

for feature in top_features:
    plt.figure(figsize=(8,4))
    sns.histplot(data=df, x=feature, hue='Class', bins=50, kde=True)
    plt.title(f"Distribution of {feature} by Class")
    plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```
